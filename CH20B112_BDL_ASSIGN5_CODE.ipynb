```python
# Load MNIST dataset
(X_train, Y_train), (X_test, Y_test) = keras.datasets.mnist.load_data()
num_classes = 10
x_train = X_train.reshape(60000, 784)
x_test = X_test.reshape(10000, 784)
x_train = x_train.astype('float32')
x_test = x_test.astype('float32')
x_train /= 255
x_test /= 255
print(x_train.shape, 'train input samples')
print(x_test.shape, 'test input samples')
```

```python
y_train = keras.utils.to_categorical(Y_train, num_classes)
y_test = keras.utils.to_categorical(Y_test, num_classes)
print(y_train.shape, 'train output samples')
print(y_test.shape, 'test output samples')
```

```python
# let's try a basic neural network for digit classification
def basic_model(epochs):
    model = Sequential()
    model.add(layers.Dense(20, activation='sigmoid', input_shape=(784,)))
    model.add(layers.Dense(20, activation='sigmoid'))
    model.add(layers.Dense(10, activation='softmax'))  
    model.summary()
    
    model.compile(loss='categorical_crossentropy', metrics=['accuracy'])
    history = model.fit(x_train, y_train, epochs = epochs, validation_data = (x_test, y_test))
    loss, accuracy = model.evaluate(x_test, y_test, verbose=0)
    model_name = "basic_model"
    
    mlflow.log_param("model_name", model_name)
    mlflow.log_param("epochs", epochs)
    mlflow.log_metric("loss", loss)
    mlflow.log_metric("accuracy", accuracy)
    mlflow.keras.log_model(model, "models", signature=signature)
    
    return history
```

```python
def basic_model_log_metrics(epochs):
    model = Sequential()
    model.add(layers.Dense(20, activation='sigmoid', input_shape=(784,)))
    model.add(layers.Dense(20, activation='sigmoid'))
    model.add(layers.Dense(10, activation='softmax'))  
    model.summary()
    
    model.compile(loss='categorical_crossentropy', metrics=['accuracy'])
    history = model.fit(x_train, y_train, epochs = epochs, validation_data = (x_test, y_test))
    loss, accuracy = model.evaluate(x_test, y_test, verbose=0)
    model_name = "basic_model"
    
    #mlflow.log_param("model_name", model_name)
    #mlflow.log_param("epochs", epochs)
    mlflow.log_metric("loss", loss)
    mlflow.log_metric("accuracy", accuracy)
    #mlflow.keras.log_model(model, "models", signature=signature)
    
    return history
```

```python
# Let's try with a slightly bigger model with more parameters
def bigger_model(epochs):
    model = Sequential()
    model.add(layers.Dense(256, activation='sigmoid', input_shape=(784,)))
    model.add(layers.Dense(128, activation='sigmoid'))
    model.add(layers.Dense(10, activation='softmax'))
    model.summary()
    
    model.compile(loss='categorical_crossentropy', metrics=['accuracy'])
    history = model.fit(x_train, y_train, epochs = epochs, validation_data = (x_test, y_test))
    loss, accuracy = model.evaluate(x_test, y_test, verbose=0)
    model_name = "bigger_model"
    
    mlflow.log_param("model_name", model_name)
    mlflow.log_param("epochs", epochs)
    mlflow.log_metric("loss", loss)
    mlflow.log_metric("accuracy", accuracy)
    mlflow.keras.log_model(model, "models", signature=signature)
    
    return history
```

```python
# let's try adding kernel regularization to the mix
def regularized_model(epochs, reg):
    model = Sequential()
    model.add(layers.Dense(256, activation='sigmoid', input_shape=(784,), kernel_regularizer=regularizers.L2(reg)))
    model.add(layers.Dense(128, activation='sigmoid', kernel_regularizer=regularizers.L2(reg)))
    model.add(layers.Dense(10, activation='softmax'))
    model.summary()
    
    model.compile(loss='categorical_crossentropy', metrics=['accuracy'])
    history = model.fit(x_train, y_train, epochs = epochs, validation_data = (x_test, y_test))
    loss, accuracy = model.evaluate(x_test, y_test, verbose=0)
    model_name = "regularized_model"
    
    mlflow.log_param("model_name", model_name)
    mlflow.log_param("epochs", epochs)
    mlflow.log_param("regularization_strength", reg)
    mlflow.log_metric("loss", loss)
    mlflow.log_metric("accuracy", accuracy)
    mlflow.keras.log_model(model, "models", signature=signature)
    
    return history
```

```python
# Dropout
def dropout_model(epochs, p1, p2):
    model = Sequential()
    model.add(layers.Dense(256, activation='sigmoid', input_shape=(784,)))
    model.add(Dropout(p1))
    model.add(layers.Dense(128, activation='sigmoid'))
    model.add(Dropout(p2))
    model.add(layers.Dense(10, activation='softmax'))
    model.summary()
    
    model.compile(loss='categorical_crossentropy', metrics=['accuracy'])
    history = model.fit(x_train, y_train, epochs = epochs, validation_data = (x_test, y_test))
    loss, accuracy = model.evaluate(x_test, y_test, verbose=0)
    model_name = "dropout_model"
 
    mlflow.log_param("model_name", model_name)
    mlflow.log_param("epochs", epochs)
    mlflow.log_param("layer1_dropout", p1)
    mlflow.log_param("layer2_dropout", p2)
    mlflow.log_metric("loss", loss)
    mlflow.log_metric("accuracy", accuracy)
    mlflow.keras.log_model(model, "models", signature=signature)
    
    return history
```

```python
# Early Stopping
def early_stopping_model(epochs, min_delta, patience):
    model = Sequential()
    model.add(layers.Dense(256, activation='sigmoid', input_shape=(784,)))
    model.add(layers.Dense(128, activation='sigmoid'))
    model.add(layers.Dense(10, activation='softmax'))
    model.summary()
   
    model.compile(loss='categorical_crossentropy', metrics=['accuracy'])
    es = keras.callbacks.EarlyStopping(monitor='val_accuracy', min_delta=min_delta, patience=patience)
    history = model.fit(x_train, y_train, epochs = epochs, validation_data = (x_test, y_test), callbacks = [es])
    loss, accuracy = model.evaluate(x_test, y_test, verbose=0)
    model_name = "early_stopping_model"
   
    mlflow.log_param("model_name", model_name)
    mlflow.log_param("epochs", epochs)
    mlflow.log_param("min_delta", min_delta)
    mlflow.log_param("patience", patience)
    mlflow.log_metric("loss", loss)
    mlflow.log_metric("accuracy", accuracy)
    mlflow.keras.log_model(model, "models", signature=signature)
       
    return history
```

```python
# Using learning rate with momentum
def model_with_lr_and_momentum(model_name, epochs, lr, **kwargs):
    model = Sequential()
    model.add(layers.Dense(20, activation='sigmoid', input_shape=(784,)))
    model.add(layers.Dense(10, activation='sigmoid'))
    model.add(layers.Dense(10, activation='softmax'))
    model.summary()
    
    optimizer = keras.optimizers.SGD(learning_rate=lr, **kwargs)
    model.compile(optimizer=optimizer, loss='categorical_crossentropy', metrics=['accuracy'])
    history = model.fit(x_train, y_train, epochs = epochs, validation_data = (x_test, y_test))
    loss, accuracy = model.evaluate(x_test, y_test, verbose=0)
    
    mlflow.log_param("model_name", model_name)
    mlflow.log_param("epochs", epochs)
    mlflow.log_param("learning rate", lr)
    for key, value in kwargs.items():
        mlflow.log_param(f"{key}", value)
    mlflow.log_metric("loss", loss)
    mlflow.log_metric("accuracy", accuracy)
    mlflow.keras.log_model(model, "models", signature=signature)
    
    return history
```

```python
# Mini-batch SGD
def mini_batch_SGD(epochs, lr, m, batch_size):
    model = Sequential()
    model.add(layers.Dense(20, activation='sigmoid', input_shape=(784,)))
    model.add(layers.Dense(10, activation='sigmoid'))
    model.add(layers.Dense(10, activation='softmax'))
    model.summary()
    
    optimizer = keras.optimizers.SGD(learning_rate=lr, momentum = m)
    model.compile(optimizer=optimizer, loss='categorical_crossentropy', metrics=['accuracy'])
    history = model.fit(x_train, y_train, epochs = epochs, validation_data = (x_test, y_test), batch_size = batch_size)
    loss, accuracy = model.evaluate(x_test, y_test, verbose=0)
    model_name = "mini_batch_SGD"
    
    mlflow.log_param("model_name", model_name)
    mlflow.log_param("epochs", epochs)
    mlflow.log_param("learning rate", lr)
    mlflow.log_param("momentum", m)
    mlflow.log_param("batch_size", batch_size)
    mlflow.log_metric("loss", loss)
    mlflow.log_metric("accuracy", accuracy)
    mlflow.keras.log_model(model, "models", signature=signature)
    
    return history
```

```python
# Mini-batch SGD with autolog
def mini_batch_SGD_with_autolog(epochs, lr, m, batch_size):
    model = Sequential()
    model.add(layers.Dense(20, activation='sigmoid', input_shape=(784,)))
    model.add(layers.Dense(10, activation='sigmoid'))
    model.add(layers.Dense(10, activation='softmax'))
    model.summary()
    
    optimizer = keras.optimizers.SGD(learning_rate=lr, momentum = m)
    model.compile(optimizer=optimizer, loss='categorical_crossentropy', metrics=['accuracy'])
    history = model.fit(x_train, y_train, epochs = epochs, validation_data = (x_test, y_test), batch_size = batch_size)
    loss, accuracy = model.evaluate(x_test, y_test, verbose=0)
    model_name = "mini_batch_SGD"
    
    mlflow.autolog()
    
    return history
```
